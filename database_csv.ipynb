```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '4' 

import deepchem as dc
import pandas as pd
import pickle
import numpy as np
import warnings

from utils import *
warnings.filterwarnings('ignore')
```

```python
freeSolve = pickle.load(open('dicts/consol.pickle', 'rb')) 

expt, tip, smiles, gbn, igb, asc, cha,nul,mobley,iupac = [], [], [], [], [], [], [], [], [],[]
for i in freeSolve.keys():
    expt.append(freeSolve[i]['expt'])
    smiles.append(freeSolve[i]['smiles'])
    tip.append(freeSolve[i]['calc'])
    gbn.append(freeSolve[i]['gbnsr6'])
    igb.append(freeSolve[i]['igb5'])
    asc.append(freeSolve[i]['asc'])
    cha.append(freeSolve[i]['cha'])
    mobley.append(i)
    iupac.append(freeSolve[i]['iupac'])
    nul.append(0)
feats = {'tip3p' : tip, 'gbnsr6' : gbn, 'igb5' : igb, 'asc' : asc, 
         'null' : nul,'chagb' : cha}
# freeSolve
```

```python
df = pd.DataFrame()
df['id'] = mobley
df['smiles'] = smiles
df['iupac name'] = iupac
df['expt'] = expt
df['tip3p'] = tip
df['cha'] = cha
df['gbnsr6'] = gbn
df['asc'] = asc
df['igb5'] = igb
df['null'] = nul
df
```

```python
df_test = df[df.index.isin(val)]
df_train = df[df.index.isin(b)]
df['train'] = df.index.isin(b)
df[df.train==0]
# rmsd(df_test['expt'],df_test['gbnsr6'])
```
